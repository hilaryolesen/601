```python
#Task 1 (and part of 2)
#Following code taken from: https://stackabuse.com/reading-and-writing-xml-files-in-python/
# and https://www.oreilly.com/library/view/python-cookbook/0596001673/ch04s11.html

import zipfile
import xml.etree.ElementTree as ET

z = zipfile.ZipFile("xml_containing_html.xml.zip", "r")

for filename in z.namelist(  ):
    unzipped=filename

tree = ET.parse(unzipped)
root = tree.getroot()

html=""
for elem in root:
    html=html+" "+elem.text
    print(elem.text)
    for subelem in elem:
        html=html+" "+subelem.text
        print(subelem.text)
```

```python
#Task 2

html = html.replace('\n', ' ').replace('\t', ' ')
html2 = "<HTML>"+html+"</HTML>"
html2 = "<HTML>"+html2+"</HTML>"
html2
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html2, "html")
print(soup.prettify())
```
